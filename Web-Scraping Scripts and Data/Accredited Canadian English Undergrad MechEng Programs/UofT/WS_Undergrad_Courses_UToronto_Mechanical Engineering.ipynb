```python
import requests as rs
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
```

```python
url1 = 'https://engineering.calendar.utoronto.ca/search-courses?keyword=&field_section_value=Mechanical%20and%20Industrial%20Engineering&field_subject_area_target_id=All&page=0'
```

```python
uClient = uReq(url1)
page_html = uClient.read()
uClient.close()
page_soup = soup(page_html, "html.parser")
```

```python
containers = page_soup.findAll("div",{"class":"view-content"})
print(len(containers))
```

```python
container_List = containers[0]
print(soup.prettify(container_List))
```

```python
course_names = container_List.findAll("p", {"class":"js-views-accordion-group-header"})
print(len(course_names))
course_names[0].text
```

```python
course_number = course_names[0].text.split(":")[0]
print(course_number)
course_name = course_names[0].text.split(":")[1:]
course_name = ':'.join([str(elem) for elem in course_name])
print(course_name) 
```

```python
containers_text = page_soup.findAll("div",{"class":"views-field views-field-field-desc"})
print(len(containers_text))
```

```python
print(containers_text[0].text)
```

```python
print("Course Number:",course_number)
print("Course Name:", course_name)
print("Course Description:",containers_text[0].text)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(columns =["Course Number","Course Name","Course Description"])
```

```python
for i in range (len(course_names)):
    course_number = course_names[i].text.split(":")[0]
    course_name = course_names[i].text.split(":")[1:]
    course_name = ':'.join([str(elem) for elem in course_name])
    course_description = containers_text[i].text
    new = {"Course Number":course_number,"Course Name":course_name,"Course Description":course_description}
    df = df.append(new,ignore_index=True)
```
