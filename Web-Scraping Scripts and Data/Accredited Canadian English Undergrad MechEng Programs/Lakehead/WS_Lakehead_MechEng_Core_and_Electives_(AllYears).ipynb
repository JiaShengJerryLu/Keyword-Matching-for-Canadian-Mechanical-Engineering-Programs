# 3. Test run successful. Implement Automation script

```python
len(list(dict.fromkeys(link_texts)))
```

```python
link_texts = list(dict.fromkeys(link_texts))
```

```python
from selenium.common.exceptions import NoSuchElementException

course_codes = []
course_names = []
course_descs = []
counter = 0

for link_text in link_texts:
    
    #go to course page
    try:
        link = driver.find_element_by_link_text(link_text)
    except NoSuchElementException:
        print("no link for {}".format(link_text))
        continue
        
    time.sleep(1)
    link.click()
    time.sleep(3)
    course_page = soup(driver.page_source, 'lxml')
    
    #scrape data
    course_codes.append(course_page.find("span", {"class": "course-code"}).text.strip())
    course_names.append(course_page.find("span", {"class": "course-title"}).text.strip())
    course_descs.append(course_page.find("span", {"class": "course-value"}).text.strip())
    
    print("Scraped ", course_codes[-1])
    counter += 1
    
    driver.back()
    time.sleep(3)
    
print("Finished scraping {} courses".format(counter))
```
