```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as ureq
from selenium import webdriver
import time
import re
```

```python
url = "https://calendar.ontariotechu.ca/preview_program.php?catoid=55&poid=10868&returnto=2434"
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--ignore-certificate-errors')
chrome_options.add_argument('--incognito')
#chrome_options.add_argument('--headless')

driver = webdriver.Chrome("C:\\Users\\jerry\\Downloads\\chromedriver", options=chrome_options)
```

```python
driver.get(url)
```

# 1. Collect course link texts for driver to click on

```python
page_soup = soup(driver.page_source, 'lxml')
```

```python
containers = page_soup.findAll("div", {"class": "acalog-core"})[5:10]
containers
```

```python
lists_of_links = [container.findAll("a") for container in containers]
link_texts = [link.text for list_of_links in lists_of_links for link in list_of_links]
link_texts
```

```python
#take out empty strings and the co-op program course
link_texts = [link.strip() for link in link_texts if link != "" and "ENGR 0998U – Engineering Internship Program" not in link]
link_texts
```

```python
len(link_texts)
```

# 2. Script to click on each link to open all course info boxes

```python
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys

for link_text in link_texts:
    link = driver.find_element_by_link_text(link_text)
        
    time.sleep(2)
    link.send_keys(Keys.DOWN)
    time.sleep(2)
    link.click()    
    time.sleep(3)
    print("clicked {}".format(link_text))
```

# 3. Obtain the updated page html and scrape all course codes, names, and descriptions

```python
page_soup = soup(driver.page_source, 'lxml')
```

```python
containers = page_soup.findAll("li", {"class": "acalog-course acalog-course-open"})
len(containers)
```

```python
course_codes = [text.split(" – ")[0] for text in link_texts]
course_codes
```

```python
course_names = [text.split(" – ")[1] for text in link_texts]
course_names
```

```python
course_descs = [cont.find("div", {"class": None}).text.strip() for cont in containers]
course_descs
```

```python
course_descs = [course_desc.split("  ")[1].split("Credit hours:")[0] for course_desc in course_descs]
course_descs
```

```python
course_descs = [desc.replace("\xa0", " ") for desc in course_descs]
course_descs
```

# 4. Write to CSV

```python
import pandas as pd

df = pd.DataFrame({
    "Course Number": course_codes,
    "Course Name": course_names,
    "Course Description": course_descs    
})

df
```

```python
df.to_csv('OntarioTechU_MechEng_EnergyEng_Specialization_Core_and_Electives_(AllYears).csv', index = False)
```
