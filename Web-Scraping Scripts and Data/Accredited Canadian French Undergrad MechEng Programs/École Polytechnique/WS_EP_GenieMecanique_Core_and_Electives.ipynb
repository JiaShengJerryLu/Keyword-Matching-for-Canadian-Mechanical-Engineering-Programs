# 1. Scrape all course information and process

```python
page_soup = soup(driver.page_source, 'lxml')
```

```python
containers = page_soup.findAll("div", {"class": "node--programme"})
len(containers)
```

```python
containers = page_soup.findAll("div", {"class": "field-structure"})
len(containers)
```

```python
course_title_containers = page_soup.find("div", {"class": "node--programme"}).findAll("tr", {"class": "titre"})
len(course_title_containers)
```

```python
course_codes = []

for container in course_title_containers:
    if container.find("td", {"class": "sigle"}) is not None:
        course_codes.append(container.find("td", {"class": "sigle"}).text.strip())
    else:
        print("not a course or unavailable: ", " ".join(container.text.split("\n")[0:7]))
```

```python
course_codes
```

```python
course_names = []
for container in course_title_containers:
    if container.find("td", {"class": "sigle"}) is not None:
        course_names.append(container.find("td", {"class": "titre"}).text.strip())
    else:
        print("not a course or unavailable: ", " ".join(container.text.split("\n")[0:7]))
```

```python
course_names
```

```python
len(course_names)
```

```python
len(course_codes)
```

```python
course_descs = []
course_desc_containers = page_soup.find("div", {"class": "node--programme"}).findAll("tr", {"class": "contenu"})

for container in course_desc_containers:
    course_descs.append(container.find("div", {"class": "desc"}).text.strip())
```

```python
len(course_descs)
```

```python
course_descs
```

```python
course_descs = [desc[len("Description"):].replace("\n\n\nDétails", "") for desc in course_descs]
```

```python
course_descs
```

# 2. Remove duplicates and write to CSV

```python
course_codes1 = []
course_names1 = []
course_descs1 = []
counter = 0

for i in range(len(course_codes)):
    if course_codes[i] not in course_codes1:
        course_codes1.append(course_codes[i])
        course_names1.append(course_names[i])
        course_descs1.append(course_descs[i])
        counter += 1
    else:
        print("Duplicate: ", course_names[i])
        
print(counter)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Course Number": course_codes1,
    "Course Name": course_names1,
    "Course Description": course_descs1
})

df
```

```python
df.to_csv('EP_GenieMecanique_Core_and_Electives_Courses.csv', index = False)
```
