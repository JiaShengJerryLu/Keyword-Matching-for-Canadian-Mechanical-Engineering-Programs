```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as ureq
from selenium import webdriver
import time
import re
```

```python
url = "https://www.umoncton.ca/repertoire"
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--ignore-certificate-errors')
chrome_options.add_argument('--incognito')
#chrome_options.add_argument('--headless')

driver = webdriver.Chrome("C:\\Users\\jerry\\Downloads\\chromedriver", options=chrome_options)
```

```python
driver.get(url)
```

# 1. Scrape all course information

```python
driver.find_element_by_link_text("Baccalauréat en ingénierie (génie mécanique)").click()
```

```python
driver.find_element_by_link_text("Description officielle").click()
```

```python
page_soup = soup(driver.page_source, 'lxml')
```

```python
course_title_containers = page_soup.find("div", {"id": "table_92"}).findAll("div", {"class": "large_screen_cours"})
len(course_title_containers)
```

```python
course_codes = [container.text.strip().split("\n")[0] for container in course_title_containers]
course_codes
```

```python
course_names = [container.text.strip().split("\n")[1] for container in course_title_containers]
course_names
```

```python
course_description_containers = page_soup.find("div", {"id": "table_92"}).findAll("div", {"class": "row mb-3 remove_display"})
len(course_description_containers)
```

```python
course_descs = [container.find("div", {"class": "col description_cours"}).text.strip() for container in course_description_containers]
course_descs
```

```python
len(list(dict.fromkeys(course_codes)))
```

# 2. Write to CSV

```python
import pandas as pd

df = pd.DataFrame({
    "Course Number": course_codes,
    "Course Name": course_names,
    "Course Description": course_descs
})

df
```

```python
df.to_csv('Moncton_GenieMecanique_Core_and_Electives_Courses.csv', index = False)
```
